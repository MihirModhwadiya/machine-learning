### Pra-8). Write a Pandas program to create a line plot of the opening, closing stock prices of given company between two specific dates.


```python
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('./Netflix/NFLX.csv')

ds.set_index('Date', inplace=True)

# start = '2021-02-04'
start = '2022-01-19'
end = '2022-02-04'

fl = ds.loc[start:end]

plt.figure(figsize=(12, 8))
plt.plot(fl.index, fl['Open'], label='Opening Price', color="blue")
plt.plot(fl.index, fl['Close'], label='Closing Price', color="red")

plt.xticks(rotation="vertical")
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Netflix Stock Price')
plt.legend()
plt.show()
```
