### Pra-9). Write a Pandas program to create a plot of Open, High, Low, Close, Adjusted Closing prices and Volume of given company between two specific dates.


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./Pra_8_data_sets/Netflix_stock_history.csv')

df.set_index('Date', inplace=True)

start_date = '2021-09-14'
end_date = '2021-09-24'

df = df.loc[start_date:end_date]

plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Open'], label='Open', linewidth=2)
plt.plot(df.index, df['High'], label='High', linewidth=2)
plt.plot(df.index, df['Low'], label='Low', linewidth=2)
plt.plot(df.index, df['Close'], label='Close', linewidth=2)
# plt.plot(df.index, df['Adj Close'], label='Adjusted Close', linewidth=2)

plt.title('Stock Prices')
plt.xlabel('Date')
plt.ylabel('Price/Volume')
plt.grid(True)
plt.legend()
plt.show()


plt.figure(figsize=(12, 3))
plt.plot(df.index, df['Volume'], label='Volume', linewidth=2, color='blue')

plt.title('Volume')
plt.xlabel('Date')
plt.ylabel('Price/Volume')
plt.grid(True)
plt.legend()
plt.show()
```
