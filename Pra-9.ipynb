### Pra-9). Write a Pandas program to create a plot of Open, High, Low, Close, Adjusted Closing prices and Volume of given company between two specific dates.


```python
# Name: Mihir V. Modhwadiya
# Enrollment: 216270307033
# Date: 1/09/2023

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./Netflix/NFLX.csv')

df.set_index('Date', inplace=True)

start_date = '2022-01-19'
end_date = '2022-02-04'

df = df.loc[start_date:end_date]

up = df[df.Close >= df.Open]
  
down = df[df.Close < df.Open]
  
width = 0.5
width2 = 0.1
  
plt.bar(up.index, up.Close-up.Open, width, bottom=up.Open)
plt.bar(up.index, up.High-up.Close, width2, bottom=up.Close)
plt.bar(up.index, up.Low-up.Open, width2, bottom=up.Open)

plt.bar(down.index, down.Close-down.Open, width, bottom=down.Open)
plt.bar(down.index, down.High-down.Open, width2, bottom=down.Open)
plt.bar(down.index, down.Low-down.Close, width2, bottom=down.Close)

plt.xticks(rotation='vertical')
plt.title('Stock Prices')
plt.xlabel('Date')
plt.ylabel('Price/Volume')
plt.grid(True)
plt.legend()
plt.show()
```
